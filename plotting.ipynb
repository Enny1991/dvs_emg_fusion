```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
import numpy as np
import csv
```

```python
data = np.loadtxt('sumit_latency.csv', delimiter=',')
latency = data[:, 0]
emg_m = data[:, 1]
emg_s = data[:, 2]
img_m = data[:, 3]
img_s = data[:, 4]
fus_m = data[:, 5]
fus_s = data[:, 6]
```

```python
with open('sumit_results_v1.csv', 'w') as csv_file:
    file_writer = csv.writer(csv_file, delimiter=',')
    header = ["feat", "index", "latency", "accuracy"]
    file_writer.writerow(header)
    
    # emg
    file_writer.writerow(['emg', 0, 0.01, emg_m[0] + np.sqrt(emg_s[0] / 2) + np.random.randn(1)[0]])
    file_writer.writerow(['emg', 1, 0.01, emg_m[0] - np.sqrt(emg_s[0] / 2)])
    
    file_writer.writerow(['emg', 0, 0.03, emg_m[1] + np.sqrt(emg_s[1] / 2)])
    file_writer.writerow(['emg', 1, 0.03, emg_m[1] - np.sqrt(emg_s[1] / 2)])
    
    file_writer.writerow(['emg', 0, 0.05, emg_m[2] + np.sqrt(emg_s[2] / 2)])
    file_writer.writerow(['emg', 1, 0.05, emg_m[2] - np.sqrt(emg_s[2] / 2)])
    
    file_writer.writerow(['emg', 0, 0.07, emg_m[3] + np.sqrt(emg_s[3] / 2)])
    file_writer.writerow(['emg', 1, 0.07, emg_m[3] - np.sqrt(emg_s[3] / 2)])
    
    file_writer.writerow(['emg', 0, 0.1, emg_m[4] + np.sqrt(emg_s[4] / 2)])
    file_writer.writerow(['emg', 1, 0.1, emg_m[4] - np.sqrt(emg_s[4] / 2)])
    
    file_writer.writerow(['emg', 0, 0.15, emg_m[5] + np.sqrt(emg_s[5] / 2)])
    file_writer.writerow(['emg', 1, 0.15, emg_m[5] - np.sqrt(emg_s[5] / 2)])
    
    file_writer.writerow(['emg', 0, 0.2, emg_m[6] + np.sqrt(emg_s[6] / 2)])
    file_writer.writerow(['emg', 1, 0.2, emg_m[6] - np.sqrt(emg_s[6] / 2)])
    
    # img
    file_writer.writerow(['img', 0, 0.01, img_m[0] + np.sqrt(img_s[0] / 2)])
    file_writer.writerow(['img', 1, 0.01, img_m[0] - np.sqrt(img_s[0] / 2)])
    
    file_writer.writerow(['img', 0, 0.03, img_m[1] + np.sqrt(img_s[1] / 2)])
    file_writer.writerow(['img', 1, 0.03, img_m[1] - np.sqrt(img_s[1] / 2)])
    
    file_writer.writerow(['img', 0, 0.05, img_m[2] + np.sqrt(img_s[2] / 2)])
    file_writer.writerow(['img', 1, 0.05, img_m[2] - np.sqrt(img_s[2] / 2)])
    
    file_writer.writerow(['img', 0, 0.07, img_m[3] + np.sqrt(img_s[3] / 2)])
    file_writer.writerow(['img', 1, 0.07, img_m[3] - np.sqrt(img_s[3] / 2)])
    
    file_writer.writerow(['img', 0, 0.1, img_m[4] + np.sqrt(img_s[4] / 2)])
    file_writer.writerow(['img', 1, 0.1, img_m[4] - np.sqrt(img_s[4] / 2)])
    
    file_writer.writerow(['img', 0, 0.15, img_m[5] + np.sqrt(img_s[5] / 2)])
    file_writer.writerow(['img', 1, 0.15, img_m[5] - np.sqrt(img_s[5] / 2)])
    
    file_writer.writerow(['img', 0, 0.2, img_m[6] + np.sqrt(img_s[6] / 2)])
    file_writer.writerow(['img', 1, 0.2, img_m[6] - np.sqrt(img_s[6] / 2)])
    
    # fus
    file_writer.writerow(['fus', 0, 0.01, fus_m[0] + np.sqrt(fus_s[0] / 2)])
    file_writer.writerow(['fus', 1, 0.01, fus_m[0] - np.sqrt(fus_s[0] / 2)])
    
    file_writer.writerow(['fus', 0, 0.03, fus_m[1] + np.sqrt(fus_s[1] / 2)])
    file_writer.writerow(['fus', 1, 0.03, fus_m[1] - np.sqrt(fus_s[1] / 2)])
    
    file_writer.writerow(['fus', 0, 0.05, fus_m[2] + np.sqrt(fus_s[2] / 2)])
    file_writer.writerow(['fus', 1, 0.05, fus_m[2] - np.sqrt(fus_s[2] / 2)])
    
    file_writer.writerow(['fus', 0, 0.07, fus_m[3] + np.sqrt(fus_s[3] / 2)])
    file_writer.writerow(['fus', 1, 0.07, fus_m[3] - np.sqrt(fus_s[3] / 2)])
    
    file_writer.writerow(['fus', 0, 0.1, fus_m[4] + np.sqrt(fus_s[4] / 2)])
    file_writer.writerow(['fus', 1, 0.1, fus_m[4] - np.sqrt(fus_s[4] / 2)])
    
    file_writer.writerow(['fus', 0, 0.15, fus_m[5] + np.sqrt(fus_s[5] / 2)])
    file_writer.writerow(['fus', 1, 0.15, fus_m[5] - np.sqrt(fus_s[5] / 2)])
    
    file_writer.writerow(['fus', 0, 0.2, fus_m[6] + np.sqrt(fus_s[6] / 2)])
    file_writer.writerow(['fus', 1, 0.2, fus_m[6] - np.sqrt(fus_s[6] / 2)])
```

```python
A = sns.color_palette("Blues", 10)
my_blues = sns.color_palette([A[-5]] + [A[-3]] + [A[-1]])
A = sns.color_palette("Reds", 10)
my_reds = sns.color_palette([A[-5]] + [A[-3]] + [A[-1]])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,8))
sns.set(font_scale=2)  # crazy big
sns.set_style("whitegrid")

data = pd.read_csv('./full_results_cross_latency_v2.csv')
data = data[data['model'] == 'sumit']
kk = sns.lineplot(data=data, x='frame_len', y='accuracy', hue='feat', style='feat', markers=True, 
             palette=my_blues, ax=ax)
print(kk)
data = pd.read_csv('sumit_results_v1.csv')
sns.lineplot(data=data, x='latency', y='accuracy', hue='feat', style='feat', markers=True, ax=ax, palette=my_reds)

ax.set_ylabel("Accuracy (%)")
ax.set_xlabel("Latency (ms)")
ax.set_xticks([0.025, 0.05, 0.075, 0.100, 0.125, 0.150, 0.175, 0.2])
ax.set_xticklabels([25, 50, 75, 100, 125, 150, 175, 200])
handles, labels = ax.get_legend_handles_labels()
labels = ["EMG", "APS", "FUS-APS", "EMG (spikes)", "DVS", "FUS-DVS"]
ax.legend(title="", loc="lower right", ncol=2, handles=handles[1:4] + handles[5:], labels=labels)
plt.savefig('latency_accuracy_cnn_v1.pdf', transparent=True, bbox_to_inches='tight')
```
