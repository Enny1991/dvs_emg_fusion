```python
plt.figure(figsize=(10, 5))
a = np.loadtxt('sumit_fus.log')[33:]
plt.plot(a[33:])
plt.plot(np.arange(20) * 38, np.ones((20)) * 1300, 'o')
plt.ylim([1000, 1500])
plt.xticks(np.arange(20) * 38)
plt.grid()
```

```python
powers = np.array([1289, 1376, 1251, 1301, 1375, 1256, 1334, 1463, 1372]) - 1085  # mW
latencies = np.array([0.018777847290039062, 
                    0.023101806640625, 
                    0.026050567626953125, 
                    0.02321767807006836, 
                    0.025551319122314453, 
                    0.024304628372192383, 
                    0.027482986450195312, 
                    0.023222923278808594, 
                    0.02789473533630371])  # s

fus_energy = powers * 0.1 * 1000
fus_edp = powers * 0.1 * latencies * 1000

print(f"LAT: {np.mean(latencies * 1000):.2f}  ms")
print(f"ENE: {np.mean(fus_energy):.2f} uJ")
print(f"EDP: {np.mean(fus_edp):.2f} uJ * s")
```
